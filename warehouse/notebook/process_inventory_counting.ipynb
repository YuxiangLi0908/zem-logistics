```python
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_excel("./data/sav_counting_1209.xlsx")
df = df.dropna()
# df = df.drop_duplicates()
df.head()
```

```python
for idx, row in df.iterrows():
    label = row["pallet_label"]
    if "|" in label:
        label = label.split("|")
    elif " " in label:
        label = label.split(" ")
    else:
        print(idx, row["pallet_label"], label)
        
    try:
        container, destination = label[0], label[1]
        destination = destination.split("-")[0]
        df.loc[idx, "container"] = container.upper()
        df.loc[idx, "destination"] = destination.upper()
    except:
        print(idx, row["pallet_label"], label)
```

```python
# df[df.pallet_label.str.contains("ZCSU7312696")].sort_values("pallet_label")
```

```python
df_agg = df.groupby(["container", "destination"]).pallet_label.count().reset_index()
df_agg.sort_values("pallet_label", ascending=False).to_csv("./data/sav_counting_1209_processed.csv", index=False)
```
